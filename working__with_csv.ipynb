```python
def rename(name):
  if name =="Royal Challengers Bangalore":
    return "RCB"
  elif name =="Rajasthan Royals":
    return "RR"
  elif name =="Sunrisers Hyderabad":
    return "SR"
  else:
    return name
  
```

```python
print(rename("Royal Challengers Bangalore"))
print(rename("Rajasthan Royals"))
rename("Sunrisers Hyderabad")
```

```python
#convertors 
df=pd.read_csv("/content/ipl-matches.csv",converters={"Team1":rename,"TossWinner":rename})
df.head(5)
```
